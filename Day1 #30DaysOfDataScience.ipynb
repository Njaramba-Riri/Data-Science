```python
x=np.linspace(0,5,200)
y1=np.sqrt(x)
y2=np.sin(x)
plt.subplot(212)
plt.plot(x,y1,'r-')
plt.subplot(211)
plt.plot(x,y2,'b--')
plt.show()
```

```python
plt.figure(1)
plt.plot(x,y1,'r-')
plt.title("Fig 1")
plt.figure(2)
plt.plot(x,y2,'g--')
plt.title("Fig 2")
plt.show()
```

```python
#Histogram
```

```python
mu,sigma=172, 4
x=mu+sigma*np.random.randn(10000)
plt.hist(x,100,density=True,facecolor="green")
plt.xlabel("Height")
plt.ylabel("Probability")
plt.title("Height Of Students")
plt.suptitle("Technical University of Kenya")
plt.text(155,0.125,"mu=172, sigma=4")
plt.axis([155,190,0,0.15])
plt.show()
```

```python
#Barchart
```

```python
kim=(90,50,45,84)
Njeri=(45,70,65,29)
Anyango=(80,60,34,55)

skills=("Python","JavaScript","SQL","Machine Learning")

width=0.2
index=np.arange(4)
plt.bar(index,kim,
       width=width,label="kim")
plt.bar(index+width,Njeri,
        width=width, label="Njeri")
plt.bar(index+width*2,Anyango,
       width=width, label="Anyango")

plt.xticks(index+width,skills)
plt.ylim(0,100)
plt.suptitle("S.C.I.T")
plt.title("IT Students Skill level")
plt.xlabel("IT Skill")
plt.ylabel("Skill Level %")
plt.legend(loc='upper center')
```

```python
#surfaceplots
```

```python
ax=plt.axes(projection='3d')

def z_funtion(x,y):
    return np.sin(np.sqrt(x**2+y**2))

x=np.linspace(-5,5,50)
y=np.linspace(-5,5,50)

X,Y=np.meshgrid(x,y)
Z=z_funtion(X,Y)
ax.plot_surface(X,Y,Z)
plt.show()
```
